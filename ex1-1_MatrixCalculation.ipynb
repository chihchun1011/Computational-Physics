```python
import numpy as np
import random
from random import uniform
```

## 1. Creates a 6x6 matrix

```python
A = np.empty((6,6))
random.seed(3227)

for i in range(A.shape[0]):

  for j in range(A.shape[1]):

    A[i][j] = uniform(-5.0,5.0)

print(A)
print(A[1][2])
```

## 2. Largest element in a matrix

```python
max = -999
idx = [-1,-1]

for i in range(A.shape[0]):

  for j in range(A.shape[1]):

    if A[i][j] > max :
      max = A[i][j]
      idx = [i,j]

print(max)
print(idx)

## compare with numpy function
print(np.max(A))
print(np.argmax(A))
```

## 3. Largest elements row and column

```python
row = np.empty(A.shape[0])

for i in range(A.shape[0]):

  row[i] = -999

  for j in range(A.shape[1]):

    if A[i][j] > row[i] : row[i] = A[i][j]

print(row)
```

```python
column = np.empty(A.shape[1])

for i in range(A.shape[1]):

  column[i] = -999

  for j in range(A.shape[0]):

    if A[j][i] > column[i] : column[i] = A[j][i]

print(column)
```

```python
## compare with np function
print(np.max(A,axis=1))
print(np.max(A,axis=0))
```

### Multiply

```python
ans = 0

for i in range(row.shape[0]):

  ans += row[i] * column[i]

print(ans)

## compare with numpy function
row_transpose = row.reshape((1,row.shape[0]))
print(np.matmul(row,column))
```
